<h1>Agrupamento com Fuzzy c-means 
<h3>Conhecido como agrupamento parcial difuso, apresenta a probabilidade das instâncias pertencerem a cada grupo

<h3>Importando as bibliotecas

```python
from sklearn import datasets #datasets
import numpy as np #para cálculos matemáticos
from sklearn.metrics import confusion_matrix #matrix de confusão

#Instalar o scikit-fuzzy pelo terminal
import skfuzzy #algoritmo de agrupamento
```

<h3>Carregando os dados

```python
iris = datasets.load_iris()
iris
```

<h3>Aplicando o algortimo

```python
#Traz os dados que vão estar utilizando do dataset iris
iris.data
```

```python
#Faço a matrix transposta
iris.data.T
```

<h4>Importante: Quando trabalho com este algoritmo, a matrix de dados que vou colocar no parametro data deve ser transposta

```python
#Aplico a matrix transposta no meu algortimo. Crio o objeto r (resultado)
#Parametros:c = Numero de clusters
#           m = Membership, indica qual grupo cada cluster pertence (na documentação é indico usar 2)
#           error = Erro. Critério de parada (na documentação é inidicado usar 0.005)
#           maxiter = Quantas vezes ele vai fazer as repetições
#           init = None
r = skfuzzy.cmeans(data = iris.data.T, c = 3, m = 2, error = 0.005, maxiter = 1000, init = None)
r
```

<h3>Separando o valores de probabilidade

```python
previsoes_porcentagem = r[1]
previsoes_porcentagem
```

```python
#Posso verificar qual a probalidade do primeiro registro (instância) pertencer aos grupo 0,1 e 2
print(previsoes_porcentagem[0][0])
print(previsoes_porcentagem[1][0])
print(previsoes_porcentagem[2][0])
#De acordo com o resultado, verifico que a probabilidade deste registro pertencer apenas ao grupo 1 é muito grande sendo de 99%
```

<h3>Vou considerar que o registro pertencerá ao grupo que apresentar uma maior probabilidade

```python
#Crio meu objeto previsoes. Utilizo a função argmax para pegar o maior valor e defino o parametro axis=0 para pegar as colunas
previsoes = previsoes_porcentagem.argmax(axis = 0)
previsoes
```

<h3>Matrix de Confusão

```python
resultados = confusion_matrix(iris.target, previsoes)
resultados
```
